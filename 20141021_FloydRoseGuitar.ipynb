## Computing linear masses from the data sheet

In this section, we're going to compute the linear weights of each individual string using the factory data that can be found on the string package.

In my case, I'm using a d'Addario EXL116 (medium top, heavy bottom) string set, whose information can be found [here](http://www.daddario.com/DADProductDetail.Page?ActiveID=3769&productid=25&productname=EXL116_Nickel_Wound__Medium_Top_Heavy_Bottom__11_52). It looks like this:

```python
from IPython.display import Image
```

```python
Image(url="http://www.daddario.com/resources/jdcdad/images/products/da_prod_exl116_detail2_1.jpg")
```

The classical formula relating the fundamental parameters of a vibrating string is:

$$
f = \frac{1}{2 L} \sqrt{\frac{T}{\mu}}
$$

From this formula, we can compute the value of $\mu$ as:

$$
\mu = \frac{T}{4 L^2 f^2}
$$

Using the frequencies found on [Wikipedia](http://en.wikipedia.org/wiki/Piano_key_frequencies) and the data on the package, we can compute $\mu$:

```python
freqs = array([82.4, 110., 146.8, 196., 246.9, 329.6]) # frequencies of the guitar strings, from low E to high E
calibration_T = array([9.59, 11.61, 11.22, 8.43, 8.09, 8.9]) # calibration tensions found on package
mu = calibration_T / (4 * 0.648**2 * freqs**2) 
```

```python
mu
```

We can plot these linear masses as a function of string number:

```python
plot(arange(1, 7), mu * 1000, 'o')
xlabel('string number')
ylabel('linear mass (g/m)')
grid(True)
xlim(0.5, 6.5)
```

## Extracting frequency information from a recorded guitar sound

In this paragraph, we will develop tools to measure the frequencies of the different guitar strings from a recorded sound.

First, let's record our guitar sound (which we have plugged into an amp). We will use code inspired by the "getting started" section of the `pyaudio` library found [here](http://people.csail.mit.edu/hubert/pyaudio/#examples). I modified the example to record a sound with the following properties:

- mono (1 channel only, to save some space)
- sample rate of 44100 kHz


```python
import pyaudio
```

```python
import pyaudio
import wave
```

```python
CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100
RECORD_SECONDS = 5
WAVE_OUTPUT_FILENAME = "files/floyd_rose_tuning1.wav"

p = pyaudio.PyAudio()

stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)

print("* recording")

frames = []

for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = stream.read(CHUNK)
    frames.append(data)

print("* done recording")

stream.stop_stream()
stream.close()
p.terminate()

wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
wf.setnchannels(CHANNELS)
wf.setsampwidth(p.get_sample_size(FORMAT))
wf.setframerate(RATE)
wf.writeframes(b''.join(frames))
wf.close()
```

Having a $\delta f$ equal to 0.2 Hz allows us to do some very precise frequency detection !

From this data, we're going to extract the frequencies of the spectrum using the following algorithm.

- find highest amplitude frequency
    - eliminate all harmonics with a given bandwidht
- repeat the above
    - if the found frequency is a divisor of the already found frequencies, eliminate harmonic
- stop when 6 different frequencies have been found

```python
def find_frequencies(n, wave_snd, bw=10, max_freq=5000):
    """finds n frequencies in the spectrum
    wave_snd is a wave file result from scipy.io.wavread
    bw is the bandwith around which the filtering is done
    """
    spectrum = fft(wave_snd[1])
    freq_vector = fftfreq(wave_snd[1].size, 1./wave_snd[0])
    
    freq_range = (freq_vector > 0.) & (freq_vector < max_freq)
    spectrum = spectrum[freq_range]
    freq_vector = freq_vector[freq_range]
    
    fundamentals = []
    while len(fundamentals) < n:
        # find highest amplitude frequency
        ind = argmax(abs(spectrum))
        new_freq = freq_vector[ind]
        
        # eliminate harmonics
        freqs_to_kill = [i*new_freq for i in arange(int(max_freq / new_freq))] 
        for f in freqs_to_kill:
            ind_low = abs(freq_vector - (f - bw / 2.)).argmin()
            ind_high = abs(freq_vector - (f + bw / 2.)).argmin()
            spectrum[ind_low:ind_high] = 0
        
        # check if found frequency is lower harmonic (i.e. duplicate)
        duplicate = False
        for f in fundamentals:
            if allclose(f / new_freq, int(f / new_freq), atol=0.01):
                fundamentals.remove(f)
                fundamentals.append(new_freq)
                duplicate = True
                print "duplicate found! new is {0:.2f}, old is {1:.2f}".format(new_freq, f)
                break
        
        # add to fundamentals
        if not duplicate:
            fundamentals.append(new_freq)
    
    return fundamentals
        
            
```

## Computing the string tensions from the frequencies

From the above steps, we know the linear masses and the frequencies of our strings. Assuming I'm tuned from low to high (which I am), we can derive the tensions and compute the total tension on the guitar neck from our data:

$$
T =  4 \mu L^2 f^2
$$

```python
def compute_tensions(freqs, linear_masses, l):
    return 4 * linear_masses * l**2 * freqs ** 2
```

```python
compute_tensions(array(fundamentals), mu, 0.7)
```

I chose 0.7 m as the string length, while in reality I hadn't screwed my neck tight so strings have in fact different lengths (the shortest being the lowest E string).

The tensions are, of course, in Newtons. So if we want something more tangible, we can convert it to kg with the following code: 

```python
T_kg = compute_tensions(array(fundamentals), mu, 0.7) / 10.
T_kg
```

The total tension in my case is:

```python
sum(T_kg)
```
