# Plotting the bounds for the dating pool as a function of your age

What's interesting here is that if you apply the creepiness rule to your own age, your get a lower bound on people you can date. But you can also look at it the other way around and ask what the upper dating age is (as suggested by the comic) where your age is the result of the creepiness rule applied to the upper dating age.
Formally:

- lower_bound = creepiness_rule(your_age)
- creepiness_rule(upper_bound) = your_age

If we denote the creepiness rule by $f$, upper and lower bounds by $u$ and $l$ and your age by $x$ then we have:

$$l = f(x)$$
$$f(u) = x \Leftrightarrow u = f^{-1}(x)$$

```python
def creepiness_rule(age):
    return age / 2. + 7

def inverse_creepiness_rule(age):
    return 2. * (age - 7)
```

```python
ages = linspace(15, 70)
plot(ages, creepiness_rule(ages), label='lower dating age')
plot(ages, inverse_creepiness_rule(ages), label='upper dating age')
legend(loc=2)
xlim((15, 70))
grid(True)
xlabel("your age (years)")
ylabel("age (years)")
```

```python
def dating_bounds(age):
    return (creepiness_rule(age), inverse_creepiness_rule(age))
```

```python
dating_bounds(18)
```

```python
dating_bounds(30)
```

Instead of importing the data from the csv file, which was quite messy, I just copy pasted the age pools from the Excel spreadsheet, after adding together the categories "Married Spouse Absent", "Widowed", "Divorced", "Separated" and "Never Married".

The data is given as the number of singles within an age category bounded by lower and upper age bounds. For example, there are 7993 thousand singles of both sexes between the ages of 18 and 19.

```python
ages_lower = array([15, 18, 20, 25, 30, 35, 40, 45, 50, 55, 65, 75, 85])
ages_upper = array([17, 19, 24, 29, 34, 39, 44, 49, 54, 64, 74, 84,  nan])
data_both_sexes = array([12740, 7993, 19063, 13970, 9233, 7277, 7403, 7971, 7708, 12733, 7825, 6206, 3428])
data_males = array([6559, 4082, 10079, 7726, 4952, 3639, 3729, 3860, 3557, 5461, 2614, 1592, 826])
data_females = array([6183, 3910, 8986, 6244, 4281, 3638, 3673, 4112, 4151, 7272, 5212, 4616, 2601])
```

As a first step, let's just plot the numbers, with the lower age bound as the x-axis value.

```python
plot(ages_lower, data_both_sexes, label="both sexes")
plot(ages_lower, data_males, label="males")
plot(ages_lower, data_females, label="females")
xlabel("age class (lower bound)")
ylabel("singles (in thousands)")
legend()
```

As one can see above, we can see two spikes that are seen in both male and female data.

- The first seems to be due to the first wave of marriages that takes place in the early twenties. 
- The second ones seems to be around the 50s and could therefore be due to the people that separate after building a family for a couple tens of years.

In fact, the curve shown in panel 3 of the xkcd comic is slightly off, as it does not show the spike seen in the real data around 50 years of age.

## Building interpolation functions on the data

Now that we have this data, we can use it to calculate the age pool as suggested by Randall Munroe. Let's start with building an interpolation function that gives the number of people above a certain age.

```python
def singles_above_age(age, data):
    if age > ages_lower[-1]:
        return data[-1]
    ind = (ages_lower >= age).nonzero()[0][0]
    singles = sum(data[ind:])
    if ind != 0: #
        delta_age = ages_lower[ind] - age
        singles += float(delta_age) / (ages_upper[ind - 1] - ages_lower[ind - 1] + 1) * data[ind - 1]
    return singles
```

Let's check if the function works.

```python
print singles_above_age(20, data_females)
print sum(data_females[2:])
```

```python
print singles_above_age(19, data_females)
print sum(data_females[2:]) + data_females[1] / 2
```

```python
print singles_above_age(18.5, data_females)
print sum(data_females[2:]) + data_females[1] / 2 * 1.5
```

```python
singles_above_age(99, data_females)
```

This seems to work. So let's build a function that returns single between two bounds.

```python
def singles_between_bounds(bounds, data):
    return singles_above_age(bounds[0], data) - singles_above_age(bounds[1], data)
```

Let's check that the function works.

```python
print singles_between_bounds((15, 18), data_males)
print data_males[0]
```

## Plotting the dating pool using the creepiness function defined in the comic

```python
def dating_pool(age, dataset):
    age_bounds = dating_bounds(age)
    return singles_between_bounds(age_bounds, dataset)
```

```python
dating_pool(26, data_females)
```

```python
dating_pool(16, data_females)
```

We can now plot the dating pool as a function of age for males and females.

```python
ages = arange(20, 80)
pool = [dating_pool(age, data_females) for age in ages]
plot(ages, pool, label='males')
pool = [dating_pool(age, data_males) for age in ages]
plot(ages, pool, label='females')
title("Dating pool")
xlabel("age (years)")
ylabel("thousands of persons")
legend(loc=2)
grid(True)
xlim((20, 70))
```

So, to comment on the comic's conclusion: indeed, the dating pool grows for men and women until ages 50 and 40, respectively. However, this assumes a creepy creepiness function. What if instead we assumed a more "normal" creepiness function, say +- 5 years, what does this look like?

```python
def dating_pool(age, dataset):
    age_bounds = (age - 5, age + 5)
    return singles_between_bounds(age_bounds, dataset)

ages = arange(20, 80)
pool = [dating_pool(age, data_females) for age in ages]
plot(ages, pool, label='males')
pool = [dating_pool(age, data_males) for age in ages]
plot(ages, pool, label='females')
title("Dating pool")
xlabel("age (years)")
ylabel("thousands of persons")
legend(loc=2)
grid(True)
xlim((20, 70))
```

So here we have it: if you're not dating at the edge of the bell curve (in terms of age), your dating pool does get down with age, even though there is a slight rebound around the 50s.
